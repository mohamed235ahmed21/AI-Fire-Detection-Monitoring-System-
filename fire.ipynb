```python

from ultralytics import YOLO
import cv2

model = YOLO('best.pt')

# Image inference
results = model.predict('fire_image.png', conf=0.25, save=True)
for r in results:
    print(f'Detected {len(r.boxes)} fire regions')
    for box in r.boxes:
        print(f'  conf={box.conf[0]:.2f}  bbox={box.xyxy[0].tolist()}')

# Webcam / Video inference
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break
    results = model.predict(frame, conf=0.25)
    annotated = results[0].plot()
    cv2.imshow('Fire Detection', annotated)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()

```
